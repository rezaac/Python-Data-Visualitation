Soal 1: Visualisasikan distribusi (KDE plot) dari kolom ```bill_length_mm``` dan kolom ```bill_depth_mm``` yang dikelompokkan berdasarkan kolom ```island```. Visualisasi dari ```bill_length_mm``` berada di kiri, sedangkan visualisasi ```bill_depth_mm``` berada di sebelah kanan (sub plots). Buatlah plot se aestetik mungkin. 

Berilah judul pada masing-masing sub-plot, kemudian berilah judul utama di tengah.

````python
df = sns.load_dataset("penguins")
````

````python
df = df.dropna().reset_index(drop = True)
````

````python
df
````

````python
fig, ax = plt.subplots(1,2,figsize=(30,10))

##Plot 1
BL = sns.kdeplot(data=df, x="bill_length_mm", hue = 'island', ax = ax[0], fill=True, palette="husl")
##Plot 2
BD = sns.kdeplot(data=df, x="bill_depth_mm", hue = 'island', ax = ax[1], fill=True, palette="husl")
BL.title.set_text('Distribusi Panjang Paruh Pinguin Berdasarkan Pulau (mm)')
BD.title.set_text('Distribusi Tebal Paruh Pinguin Berdasarkan Pulau (mm)')

#mengubah labels Bill Length
BL.set_title('Bill Length', fontsize=20, fontweight='bold')
BL.set_xlabel('Bill Length (mm)', fontsize=16, fontweight='normal')

#mengubah plot Bill Depth
BD = sns.kdeplot(data=df, x="bill_depth_mm", hue="island", fill=True, palette="husl", ax=ax[1])
#mengubah labels Bill Depth
BD.set_title('Bill Depth', fontsize=20, fontweight='bold')
BD.set_xlabel('Bill Depth (mm)', fontsize=16, fontweight='normal')


plt.tight_layout(pad=3.0)


fig.suptitle('KDE Plot Panjang dan Tebal Paruh Pinguin Berdasarkan Pulau', size = 20)
````

Soal 2: Visualisasikan scatterplot dengan ketentuan sebagai berikut:
- bentuk : subplot dengan 1 column dan 2 row.
- subplot 1:
    - sumbu x : body_mass_g
    - sumbu y : flipper_length_mm
- subplot 2:
    - sumbu x : body_mass_g
    - sumbu y : bill_length_mm
- diwarnai titik2nya berdasarkan island
- buat dengan aestetik
- Kemudian, tulis interpretasi yang Anda peroleh berdasarkan scatterplot tersebut.

````python
fig, ax = plt.subplots(2,1,figsize=(30,20))

##Plot flipper_length
a = sns.scatterplot(data=df, x='body_mass_g', y='flipper_length_mm', hue = 'island', ax = ax[0], s=120, style='island', palette='husl')
a.set_title('Panjang Sayap vs Berat Badan', fontsize=20, fontweight='bold')
a.set_xlabel('Berat badan (gr)', fontsize=17, fontweight='normal')
a.set_ylabel('Panjang sayap (mm)', fontsize=17, fontweight='normal')
a.grid()

##Plot bill_length
b = sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', hue = 'island', ax = ax[1], s=120, style='island', palette='husl')
b.set_title('Panjang Paruh vs Berat Badan', fontsize=20, fontweight='bold')
b.set_xlabel('Berat badan (gr)', fontsize=17, fontweight='normal')
b.set_ylabel('Panjang paruh (mm)', fontsize=17, fontweight='normal')
b.grid()
````

Interpretasi:
1. Penguin dari pulau biscoe memiliki berat badan yang lebih berat dan panjang sayap yang lebih panjang dari penguin pulau Torgersen dan pulau Dream
2. Penguin dari pulau Dream dan pulau Torgersen memiliki berat badan yang kurang lebih sama, namun mayoritas penguin pulau Dream memiliki panjang paruh yang lebih panjang dari pulau Torgersen
3. Penguin dari pulau Dream secara mayoritas memiliki panjang paruh yang lebih tinggi dari penguin jenis lain pada berat badan yang sama
